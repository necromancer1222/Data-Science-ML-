```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline
```

```python
data = np.array(pd.read_csv('data3.csv', header=None))
# Assign the features to the variable X, and the labels to the variable y. 
X=data[:,0:2]
y=data[:,2]
```

```python
# dplit data into train and test data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
#Create the decision tree model and assign it to the variable model.
model = DecisionTreeClassifier()
```

```python
#Fit the model to the training data.
model.fit(X_train,y_train)
```

```python
#Make predictions on the test data by the model you have traind
y_pred = model.predict(X_test)
```

```python
#Calculate the accuracy and assign it to the variable acc. on the test data
acc = accuracy_score(y_test, y_pred)
acc
```

```python
y_pred_t = model.predict(X_train)
```

```python
acc = accuracy_score(y_train, y_pred_t)
acc
```
