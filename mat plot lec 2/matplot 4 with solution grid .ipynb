# Histogram

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline
```

```python
pokemon=pd.read_csv('pokemon.csv')
print("shape =",pokemon.shape)
pokemon.head(10)
```

# Creating histogram by matplot

```python
# plt.hist(data = df, x = 'num_var') in ten bins
plt.hist(data=pokemon,x='speed')
```

```python
# plt.hist(data = df, x = 'num_var',bins=20) 20 bins
plt.hist(data=pokemon,x='speed',bins=20)
```

# How to get the best bins value ?

```python
bins=np.arange(0,pokemon['speed'].max()+5,5)
plt.hist(data=pokemon,x='speed',bins=bins)
```

```python
# sb.distplot(df['num_var'],kde=False)
sb.distplot(pokemon['speed'],kde=False)
```

# create two figures

```python
plt.figure(figsize = [10, 5])
# example of somewhat too-large bin size
plt.subplot(1, 2, 1) # 1 row, 2 cols, subplot 1
bins=np.arange(0,pokemon['speed'].max()+5,5)
plt.hist(data=pokemon,x='speed',bins=bins)

# example of somewhat too-small bin size
plt.subplot(1, 2, 2) # 1 row, 2 cols, subplot 2
bins=np.arange(0,pokemon['speed'].max()+2,2)
plt.hist(data=pokemon,x='speed',bins=bins);

```

```python
bins=np.arange(0,pokemon['speed'].max()+5,5)

fig, axes = plt.subplots(3, 4) # grid of 3x4 subplots
#print(fig)
#print(axes)
axes = axes.flatten() # reshape from 3x4 array into 12-element vector
for i in range(12):
    axes[i].hist(data=pokemon,x='speed',bins=bins);
```

```python

# example of somewhat too-large
fig, axes = plt.subplots(4, 3,figsize = [20, 20]) # grid of 3x4 subplots
#print(axes)
axes = axes.flatten() # reshape from 3x4 array into 12-element vector
for i in range(12):
    bins=np.arange(0,pokemon['speed'].max()+1+i,1+i)
    axes[i].hist(data=pokemon,x='speed',bins=bins)
```

```python
plt.figure(figsize = [20, 20])
# example of somewhat too-large bin size
for i in range(4):
    plt.subplot(2, 2, 1+i) # 1 row, 2 cols, subplot 1
    bins=np.arange(0,pokemon['speed'].max()+i+5,i+5)
    plt.hist(data=pokemon,x='speed',bins=bins)
    plt.xticks(bins);
    plt.xticks(rotation=90);
```

```python
pokemon['mean_power']=pokemon.loc[:,'hp':'special-defense'].mean(axis=1)
pokemon
```

```python

plt.figure(figsize = [20, 20])
# example of somewhat too-large bin size
j=5
for i in range(1,13,3):#i=1,4,7,10
    plt.subplot(4, 3, i) # 1 row, 2 cols, subplot 1 i=4
    bins=np.arange(0,pokemon['speed'].max()+j,j)#j=5
    plt.hist(data=pokemon,x='speed',bins=bins)
    #plt.xticks(bins);
    #plt.xticks(rotation=90);
    plt.xlabel("speed at bins width "+str(j));
    plt.subplot(4, 3, i+1) # 1 row, 2 cols, subplot 1 i=1 i+1=2
    bins=np.arange(0,pokemon['hp'].max()+j,j)
    plt.hist(data=pokemon,x='hp',bins=bins)
    #plt.xticks(bins);
    #plt.xticks(rotation=90);
    plt.xlabel("health at bins width "+str(j));
    plt.subplot(4, 3, i+2) # 1 row, 2 cols, subplot 1  i=1 i+2=3
    bins=np.arange(0,pokemon['mean_power'].max()+j,j)
    plt.hist(data=pokemon,x='mean_power',bins=bins)
    plt.xticks(bins);
    #plt.xticks(rotation=90);
    plt.xlabel("mean power at bins width "+str(j));
    j=j+2#j=7
    
    
    
```

# featrues realations
* relation between id genaration from (2:4) with respect to mean power > 50

```python
pok_rel=pokemon[  (pokemon['mean_power']>50)     &  ((pokemon["generation_id"]>=2)  &    (pokemon["generation_id"]<=4 ))  ]
pok_rel
```

# graph genearation id from 2 :4 with respect to mean power >= 50

```python
sb.countplot(data=pok_rel,x='generation_id');
plt.xlabel("genearation id from 2 :4 with respect to mean power >= 50");
```

# graph mean power >= 50  with respect to genearation id from 2 :4  

```python
bins=np.arange(0,pokemon['mean_power'].max()+5,5)
plt.hist(data=pok_rel,x='mean_power',bins=bins)
plt.xticks(bins);
plt.xticks(rotation=90);
plt.xlabel("mean power >= 50 with respect to genearation id from 2 :4");
```

# compare between two histogram in the same graph

```python
pok_rel2=pokemon[(pokemon['mean_power']>50) & (pokemon["generation_id"]==2)] #  blue
pok_rel3=pokemon[(pokemon['mean_power']>50) & (pokemon["generation_id"]==3)] #  red
bins=np.arange(0,pokemon['mean_power'].max()+5,5)
plt.hist(data=pok_rel2,x='mean_power',bins=bins,color="blue")
plt.hist(data=pok_rel3,x='mean_power',bins=bins,color="red")
plt.xticks(bins);
plt.xticks(rotation=90);
```

```python
plt.figure(figsize = [20, 20])
fet_sel=["speed","hp","mean_power"] #  index  0   1   2
fet_index=0   # represnt at the start the index of "speed"
j=5
for i in range(1,13,1):
    plt.subplot(4, 3, i) # i=3
    bins=np.arange(0,pokemon[fet_sel[fet_index]].max()+j,j) # fet_sel[fet_index]="speed"   where fet_index =0    and j = 6
    plt.hist(data=pokemon,x=fet_sel[fet_index],bins=bins);
    plt.xlabel(fet_sel[fet_index]+" at bins width "+str(j));
    fet_index+=1  # fet_index=3
    if i%3==0:
        j+=1  # j=6
        fet_index=0  # fet_index=0
```
