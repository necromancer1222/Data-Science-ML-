# Choosing a Plot for Discrete Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline
```

```python
pokemon=pd.read_csv('pokemon.csv')
print("shape =",pokemon.shape)
pokemon.head(10)
```

```python

bins=np.arange(0,pokemon['speed'].max()+10,10)
plt.hist(data=pokemon,x='speed',bins=bins);
plt.xticks(bins);


```

```python
bins=np.arange(0,pokemon['speed'].max()+10,10)
plt.hist(data=pokemon,x='speed',bins=bins,rwidth = 0.8);
plt.xticks(bins);
```
