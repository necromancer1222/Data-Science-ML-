# another way: by using math and statistics (interquartile range)

##  Q1 represnt 25% of the data
##  Q3 represnt 75% of the data 
## median represnt 50% ot the data

```python
np.sort(x)
```

```python
q1,med,q3=np.percentile(x,[25,50,75])
print("Q1=",q1)
print("Q3=",q3)
print("median=",med)
iqr=q3-q1
print("IQR=",iqr)
```

```python
mn=q1 - (1.5*iqr)
mx=q3 + (1.5*iqr)
print(mn,q1,med,q3,mx)
```

```python
outliers=x[x>mx]
```

```python
print("outliers=",outliers)
```
