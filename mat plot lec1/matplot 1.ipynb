# Bar Charts

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
pokemon=pd.read_csv('pokemon.csv')
print("shape =",pokemon.shape)
pokemon.head()
```

# Creating Bar Chart using seaborn for generation id column

```python
#sb.countplot(data = df, x = 'cat_var')

sb.countplot(data=pokemon,x='generation_id')
```

# creating Bar Chat using seaborn one color

```python
sb.countplot(data=pokemon,x='generation_id',color='orange')
```

# Choosing color from  sb palette

```python
sb.color_palette()
```

```python
clr=sb.color_palette()[2]
clr
```

```python
sb.countplot(data=pokemon,x='generation_id',color=clr)
```

# sorting data in order manually

```python
sb.countplot(data=pokemon,x='generation_id',color=clr,order=[5,1,3,4,2,7,6])
```

# sorting data in order by code

```python
# by using pandas
ord1=pokemon['generation_id'].value_counts().index
ord1
```

```python
sb.countplot(data=pokemon,x='generation_id',color=clr,order=ord1)
```

# Creating Bar Chat using seaborn for type_1 column

```python
sb.countplot(data=pokemon,x='type_1',color=clr);
plt.xticks(rotation=90);
```

```python
od = pokemon['type_1'].value_counts().index
```

```python
sb.countplot(data=pokemon,y='type_1',color=clr, order = od)
```
