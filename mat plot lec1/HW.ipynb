```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline
```

```python
pokemon=pd.read_csv('pokemon.csv')
print("shape =",pokemon.shape)
pokemon.head(10)
```

```python
#pokemon['mean_power'] = (pokemon['hp'] + pokemon['attack'] + pokemon['defense'] + pokemon['speed'])/4

pokemon['mean_power'] = pokemon.loc[:,'hp':'special-defense'].mean(axis=1) #av from rows

pokemon
```

```python
fig, axes = plt.subplots(4, 3, figsize=[20,20])
axes = axes.flatten()
c = 5

for i in range(0,12,3):
    bins = np.arange(0,pokemon['speed'].max()+c, c)
    axes[i].hist(data= pokemon, x= 'speed', bins= bins)
    
    bins = np.arange(0,pokemon['hp'].max()+c, c)
    axes[i+1].hist(data = pokemon, x= 'hp',bins=bins)
    
    bins = np.arange(0,pokemon['mean_power'].max()+c, c)
    axes[i+2].hist(data = pokemon, x = 'mean_power',bins= bins)
    
    c=c+2
```
