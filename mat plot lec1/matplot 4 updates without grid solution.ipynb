# Histogram

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline
```

```python
pokemon=pd.read_csv('pokemon.csv')
print("shape =",pokemon.shape)
pokemon.head(10)
```

# Creating histogram by matplot

```python
# plt.hist(data = df, x = 'num_var') in ten bins
plt.hist(data=pokemon,x='speed')
```

```python
# plt.hist(data = df, x = 'num_var',bins=20) 20 bins
plt.hist(data=pokemon,x='speed',bins=20)
```

# How to get the best bins value ?

```python
bins=np.arange(0,pokemon['speed'].max()+5,5)
plt.hist(data=pokemon,x='speed',bins=bins)
```

# # Creating histogram by seaborn

```python
# sb.distplot(df['num_var'])
sb.distplot(pokemon['speed'],bins=20)
```

```python
# sb.distplot(df['num_var'],kde=False)
sb.distplot(pokemon['speed'],kde=False)
```

# create two figures

```python
plt.figure(figsize = [10, 5])
# example of somewhat too-large bin size
plt.subplot(1, 2, 1) # 1 row, 2 cols, subplot 1
bins=np.arange(0,pokemon['speed'].max()+5,5)
plt.hist(data=pokemon,x='speed',bins=bins)

# example of somewhat too-small bin size
plt.subplot(1, 2, 2) # 1 row, 2 cols, subplot 2
bins=np.arange(0,pokemon['speed'].max()+2,2)
plt.hist(data=pokemon,x='speed',bins=bins);

```

```python
bins=np.arange(0,pokemon['speed'].max()+5,5)

fig, axes = plt.subplots(3, 4) # grid of 3x4 subplots+
print(fig)
print(axes)
axes = axes.flatten() # reshape from 3x4 array into 12-element vector
print(axes)
for i in range(12):
    axes[i].hist(data=pokemon,x='speed',bins=bins);
```

```python

# example of somewhat too-large
fig, axes = plt.subplots(4, 3,figsize = [20, 20]) # grid of 3x4 subplots
#print(axes)
axes = axes.flatten() # reshape from 3x4 array into 12-element vector
for i in range(12):
    bins=np.arange(0,pokemon['speed'].max()+1+i,1+i)
    axes[i].hist(data=pokemon,x='speed',bins=bins)
```

```python
plt.figure(figsize = [20, 20])
# example of somewhat too-large bin size
for i in range(4):
    plt.subplot(2, 2, 1+i) # 1 row, 2 cols, subplot 1
    bins=np.arange(0,pokemon['speed'].max()+i+5,i+5)
    plt.hist(data=pokemon,x='speed',bins=bins)
    plt.xticks(bins);
    plt.xticks(rotation=90);
```
